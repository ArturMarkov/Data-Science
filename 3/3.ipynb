As dataset I am going to use data from first lab. Also I'm gonna make a loss of values to demonstrate skill of using linear interpolation to recover missing data.


```python
import requests
from io import StringIO

def createFrame(ID):
    if type(ID) is not type(list()):
        ID = [ID]
    df = pd.DataFrame()
    
    def cleanup(data: str):
        start = data.find("<pre>") + 5
        end = data.find("</pre>")
        data = data[start:end]
        data = data.replace(" ", "")
        data = data.replace(",\\n", "\n")
        data = "Year,Week,SMN,SMT,VCI,TCI,VHI\n" + data
        return data
    
    for id in ID:
        url = f"https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={id}&year1=1981&year2=2020&type=Mean"
        with requests.get(url, headers={"Accept": "text/plain"}) as plain:
            data = str(plain.content)
            data = cleanup(data)
            data = StringIO(f"""
{data}
            """)
            tmp = pd.read_csv(data, na_values=-1)
            tmp["Province"] = id
            df = df.append(tmp)
        if len(ID) > 1:
            df = (df.reset_index()
                 .drop(["index"], axis=1))
    return df
```

### Recovering missing values:
---
Now we will try to recover missing SMT values using linear interpolation. As argument, we're gonna use SMN as it's correlation coeffitient with SMT is close to 1.

```python
missing_index = df.loc[np.isnan(df["SMT"])].index
df_without_null = df.dropna()

SMT = pd.DataFrame(np.zeros((len(missing_index), 2)), columns=["lost", "recovered"], index=missing_index)
SMT["lost"] = lost

for i in missing_index:
    smn = df_without_null.loc[:, "SMN"]
    current = df.loc[i, "SMN"]
    
    left_x = np.max(smn[smn < current])
    right_x = np.min(smn[smn > current])
    
    if np.isnan(left_x):
        left_x = right_x
        right_x = np.min(smn[smn > left_x])
    elif np.isnan(right_x):
        right_x = left_x
        left_x = np.max(smn[smn < right_x])
    
    left_y = (df_without_null[df_without_null["SMN"] == left_x]
             .iloc[0, 3])
    right_y = (df_without_null[df_without_null["SMN"] == right_x]
             .iloc[0, 3])
    
    k = (right_y - left_y) / (right_x - left_x)
    b = right_y - k * right_x

    SMT.loc[i, "recovered"] = k * current + b
    
SMT.corr()
```

### Input recovered values in dataframe:

```python
df.loc[missing_index, "SMT"] = SMT["recovered"]

df
```
